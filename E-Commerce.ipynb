```python
import pandas as pd
import nltk
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
```

```python
# Download the necessary resources if not already downloaded
nltk.download('punkt')
nltk.download('stopwords')
nltk.download('wordnet')

# Initialize the WordNet lemmatizer and stopwords
lemmatizer = WordNetLemmatizer()
stop_words = set(stopwords.words('english'))

# Function to preprocess a single title
def preprocess_title(title):
    # Tokenize the title into words
    tokens = word_tokenize(title)
    # Remove stop words and punctuation
    filtered_tokens = [token for token in tokens if token.lower() not in stop_words and token.isalnum()]
    # Lemmatize each word
    lemmatized_tokens = [lemmatizer.lemmatize(token.lower()) for token in filtered_tokens]
    return lemmatized_tokens

# Apply preprocessing to the 'title' column and create new columns for tokens, stop words removed, and lemmatized tokens
df['tokens'] = df['description'].apply(word_tokenize)
df['tokens_no_stopwords'] = df['tokens'].apply(lambda tokens: [token for token in tokens if token.lower() not in stop_words and token.isalnum()])
df['lemmatized_tokens'] = df['tokens_no_stopwords'].apply(lambda tokens: [lemmatizer.lemmatize(token.lower()) for token in tokens])
```

```python
import pandas as pd
import nltk
from nltk.sentiment.vader import SentimentIntensityAnalyzer

# Download the VADER lexicon
nltk.download('vader_lexicon')

# Assuming you have a DataFrame named 'df' with a column 'lemmatized_tokens'

# Initialize the VADER sentiment analyzer
sid = SentimentIntensityAnalyzer()

# Function to compute sentiment score for each token
def compute_sentiment_score(tokens):
    scores = []
    for token in tokens:
        score = sid.polarity_scores(token)['compound']
        scores.append(score)
    return scores

# Apply sentiment analysis to lemmatized tokens
df['sentiment_scores'] = df['lemmatized_tokens'].apply(compute_sentiment_score)

# Average sentiment score for each row
df['average_sentiment_score'] = df['sentiment_scores'].apply(lambda x: sum(x) / len(x) if len(x) > 0 else 0)

# Classify sentiment label based on average score
def classify_sentiment(score):
    if score >= 0.05:
        return 'Positive'
    elif score <= -0.05:
        return 'Negative'
    else:
        return 'Neutral'

df['sentiment_label'] = df['average_sentiment_score'].apply(classify_sentiment)
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import re
```

```python
# Count plot for discount categories
plt.figure(figsize=(30, 15))
sns.countplot(x='discount', data=df, order=df['discount'].value_counts().index)
plt.title('Frequency of Different Discount Categories')
plt.xlabel('Discount Category')
plt.ylabel('Frequency')
plt.xticks(rotation=45)
plt.show()

# Extract numerical discount values
def extract_numerical_discount(discount):
    numerical_discount = re.findall(r'\d+\%', discount)
    if numerical_discount:
        return int(numerical_discount[0][:-1])  # Extract numerical value and remove '%' symbol
    else:
        return None

# Apply numerical discount extraction
df['numerical_discount'] = df['discount'].apply(extract_numerical_discount)

# Perform numerical analysis
if 'numerical_discount' in df.columns:
    numerical_analysis = df['numerical_discount'].describe()
    print("Numerical Discount Analysis:")
    print(numerical_analysis)
else:
    print("No numerical discount values available.")
```
